```python
from datasets import load_dataset
dataset = load_dataset("birgermoell/Italian_Parkinsons_Voice_and_Speech")
```

```python
import librosa
import numpy as np
import os
from tqdm import tqdm

os.makedirs("mfcc_italian", exist_ok=True)

X_train, y_train = [], []

for i, sample in tqdm(enumerate(dataset["train"])):
    audio = sample["audio"]["array"]
    sr = sample["audio"]["sampling_rate"]
    path = sample["audio"]["path"]

    # Label bepalen op basis van directorynaam in pad
    if "Healthy" in path:
        label = 0
    elif "Parkinson" in path:
        label = 1
    else:
        continue

    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)
    X_train.append(mfcc_mean)
    y_train.append(label)

X_train = np.array(X_train)
y_train = np.array(y_train, dtype=int)

print(f"MFCC's geëxtraheerd: {X_train.shape[0]} samples")
print(f"Distributie: {np.bincount(y_train)} (Gezond, Parkinson)")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve
)
from sklearn.manifold import TSNE


# SPLIT & TRAIn

X_train_it, X_val_it, y_train_it, y_val_it = train_test_split(
    X_train, y_train, test_size=0.2, stratify=y_train, random_state=42
)

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_it, y_train_it)

#VOORSPELLING ITALIAANS
y_pred_it = clf.predict(X_val_it)
y_prob_it = clf.predict_proba(X_val_it)[:, 1]
report_it = classification_report(y_val_it, y_pred_it, output_dict=True)
auc_it = roc_auc_score(y_val_it, y_prob_it)

# VOORSPELLING ENGELS 
y_pred_en = clf.predict(X_test)
y_prob_en = clf.predict_proba(X_test)[:, 1]
report_en = classification_report(y_test, y_pred_en, output_dict=True)
auc_en = roc_auc_score(y_test, y_prob_en)

#VERGELIJKINGSTABEL
df_compare = pd.DataFrame({
    "Metric": ["Accuracy", "Precision (PD)", "Recall (PD)", "F1 (PD)", "AUC"],
    "Italiaans": [
        report_it["accuracy"],
        report_it["1"]["precision"],
        report_it["1"]["recall"],
        report_it["1"]["f1-score"],
        auc_it
    ],
    "Engels": [
        report_en["accuracy"],
        report_en["1"]["precision"],
        report_en["1"]["recall"],
        report_en["1"]["f1-score"],
        auc_en
    ]
})
display(df_compare)
```
